```python
import pandas as pd 

df_gene = pd.read_csv('data/gene.csv', index_col='patient')
df_gene.head()
```

```python
df_gene.columns 
```

```python
from sklearn.metrics import accuracy_score 
from scipy.stats import ttest_ind

X_cols = df_gene.columns.drop('cancer').drop('train').to_list()
X_cols
```

```python
import numpy as np 
def get_corr(X, y):
    stat, pval = ttest_ind(X[y == 'AML'], X[y == 'ALL'], equal_var=False)
    return np.abs(stat), pval

stat, pval = get_corr(df_gene[X_cols], df_gene['cancer'])
stat, pval
```

```python
len(stat), len(pval), len(X_cols)
```

```python
pd.concat([df_gene.iloc[:, np.argsort(pval)[:8]], df_gene['cancer']], axis=1).melt(id_vars='cancer')
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt

# pvalue 작은 순부터 -> 두 분포가 많이 다른..
sns.displot(
    data=pd.concat([df_gene.iloc[:, np.argsort(pval)[:8]], df_gene['cancer']], axis=1).melt(id_vars='cancer'),
    x='value', col='variable', hue='cancer', col_wrap=4, kind='kde'
)
plt.show()
```

```python
df_gene_train = df_gene[df_gene['train'] == True]
df_gene_test = df_gene[df_gene['train'] == False]
feat_num = np.arange(1, 201)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

lr_test_accs = []
for i in feat_num:
    model = make_pipeline(
        SelectKBest(get_corr, k=i),
        StandardScaler(),
        LogisticRegression(solver='lbfgs')
    )
    model.fit(df_gene_train[X_cols], df_gene_train['cancer'])
    test_acc = accuracy_score(df_gene_test['cancer'], model.predict(df_gene_test[X_cols]))
    lr_test_accs.append(test_acc)
    
lsvm_test_accs = []
for i in feat_num:
    model = make_pipeline(
        SelectKBest(get_corr, k=i),
        StandardScaler(),
        LinearSVC(C=1000)
    )
    model.fit(df_gene_train[X_cols], df_gene_train['cancer'])
    test_acc = accuracy_score(df_gene_test['cancer'], model.predict(df_gene_test[X_cols]))
    lsvm_test_accs.append(test_acc)
```
