```python
df_titanic = pd.read_csv('data/titanic.csv', index_col='PassengerId')
df_titanic.info()
```

### iloc
대부분은 number로 index 하고자 할 때 씀 (Purely integer-location based indexing for selection by position.)

with scalar integer

```python
df_titanic.iloc[2]
```

with a list of integers

```python
df_titanic.iloc[[1,2]]
```

```python
df_titanic.iloc[range(10)]
```

with a slice object

```python
df_titanic.iloc[:3]
```

with a boolean mask

```python
df_titanic.iloc[[True] * len(df_titanic)]
```

```python
df_titanic.iloc[[True, False, True]]
```

indexing both axes

```python
df_titanic.iloc[0, 1]
```
